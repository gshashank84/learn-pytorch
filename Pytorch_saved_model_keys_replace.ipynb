```
# ALL together
from collections import OrderedDict

def func(keys_passed):
  change = ['.0.','.1.','.2.']
  if change[0] in keys_passed:
    i = change[0]
    b = keys_passed.split(i)
    b.insert(1,i[1:])
    c = ''.join(b)
    return c
  elif change[1] in keys_passed:
    i = change[1]
    b = keys_passed.split(i)
    b.insert(1,i[1:])
    c = ''.join(b)
    return c
  elif change[2] in keys_passed:
    i = change[2]
    b = keys_passed.split(i)
    b.insert(1,i[1:])
    c = ''.join(b)
    return c
  else:
    return keys_passed  

final_dic = {func(key):value for key, value in checkpoint['state_dict'].items()}
checkpoint['state_dict'] = OrderedDict(final_dic)
```
